```python
import pandas as pd
```

```python
df = pd.read_csv('P_small2_Guncrimes_ChicagoDataPortal.csv')
```

```python
df.head(1)
```

```python
df.tail(2)
```

```python
df2 = df.drop(['ID','Date','Arrest','Location'], axis=1)
df2.head(1)
```

```python
df3 = df2.set_index(['Community Area','Year']).stack().reset_index()
df3
```

```python
df4 = df3.rename(columns={'level_2':'NA', 0:'Reported Incident'})
df4
```

```python
count_series = df4.groupby(['Community Area', 'Year','Reported Incident']).size()
count_series
```

```python
df5 = count_series.to_frame(name = 'Total for Year').reset_index()
df5
```

```python
from google.colab import files
df5.to_csv('ChicagoDataPortal22_tidy.csv') 
files.download('ChicagoDataPortal22_tidy.csv')
```
